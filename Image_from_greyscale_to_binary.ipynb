```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
grey=mpimg.imread('greyscale.jpeg')
grey_img=cv2.cvtColor(grey,cv2.COLOR_BGR2RGB)
plt.imshow(grey_img)
plt.title('greyscale image')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
# %matplotlib inline

img = cv2.imread('greyscale.jpeg',0)
print(img)
```

```python
ret,binary_img=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(binary_img)
plt.title('binary_img')
plt.show()

print(binary_img)
```
